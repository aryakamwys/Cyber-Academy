## **Cross Validation Split**

Tahapan yang dilakukan pada codelab ini sebagai berikut:
1. Impor library yang dibutuhkan.
2. Pisahkan antara atribut dan label pada dataset.
3. Buat model decision tree.
4. Hitung hasil cross validation dari model dengan fungsi cross_val_score().

```python
import sklearn
from sklearn import datasets

# Load iris dataset
iris = datasets.load_iris()
```

```python
# mendefinisikan atribut dan label pada dataset
x=iris.data
y=iris.target
```

```python
from sklearn import tree

# membuat model dengan decision tree classifier
clf = tree.DecisionTreeClassifier()
```

Setelah dataset dan model siap, kita bisa menggunakan cross validation untuk mengevaluasi performa dari model machine learning. Fungsi cross_val_score() seperti di bawah menerima 4 parameter yaitu, ‘clf’ yang merupakan model machine learning, ‘X’ yang merupakan atribut dari dataset, ‘y’ yang merupakan label dari dataset, dan ‘cv’ yang merupakan jumlah fold yang akan dipakai pada cross validation.

```python
from sklearn.model_selection import cross_val_score

# mengevaluasi performa model dengan cross_val_score
scores = cross_val_score(clf, x, y, cv=5)
print(scores)
```

Elemen pertama dari larik menunjukkan nilai 0.96666 yang berarti ketika fold pertama dijadikan validation set dan fold lainnya dijadikan train set, hasil dari pengujian tersebut adalah akurasi sebesar 0.96666.

Tahapan yang ada pada latihan ini antara lain:

Pertama kita akan mengimpor library yang dibutuhkan dan mempersiapkan dataset. Dataset dapat anda unduh di tautan berikut (https://www.kaggle.com/datasets/uciml/iris). Setelah data diunduh, masukkan berkas Iris.csv ke dalam Google Colab. Lalu jangan lupa konversi dataset menjadi Pandas dataframe.

```python
import pandas as pd

# Membaca file iris.csv
iris = pd.read_csv('Iris.csv')
```

```python
# Melihat informasi dataset
iris.info()

# melihat informasi dataset pada 5 baris pertama
iris.head()
```

```python
# menghilangkan kolom yang tidak penting
iris.drop('Id',axis=1,inplace=True)
```

Sebelum melatih model kita perlu memisahkan atribut dengan label. Selain itu, kita juga perlu membagi dataset menjadi data latih dan data uji.

```python
# memisahkan atribut dan label
X = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm' ]]
y = iris['Species']

# Membagi dataset menjadi data latih & data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)
```

Selanjutnya, definisikan model decision tree yang akan kita gunakan. Kemudian,  latih model menggunakan data latih menggunakan fungsi fit().

```python
from sklearn.tree import DecisionTreeClassifier

# membuat model Decision Tree
tree_model = DecisionTreeClassifier()

# Melatih model dengan menggunakan data latih
tree_model = tree_model.fit(X_train, y_train)
```

```python
# Evaluasi Model
from sklearn.metrics import accuracy_score

y_pred = tree_model.predict(X_test)

acc_secore = round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ', acc_secore)
```

```python
# prediksi model dengan tree_model.predict([[SepalLength, SepalWidth, PetalLength, PetalWidth]])
print(tree_model.predict([[6.2, 3.4, 5.4, 2.3]])[0])
```

## **SKLearn Linear Regression**

Berikut adalah tahapan latihan yang akan Anda lakukan:
1. Impor library yang dibutuhkan.
2. Buat dataset dummy dengan Numpy Array.
3. Buat plot dari model.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data jumlah kamar
bedrooms = np.array([1,1,2,2,3,4,4,5,5,5])

# Data harga rumah dalam dollar
house_price = np.array([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000,85000, 90000])
```

```python
# Menampilkan scatter plot dari dataset
plt.scatter(bedrooms, house_price)
plt.xlabel("Jumlah Kamar")
plt.ylabel("Harga Rumah ($)")
plt.title("Hubungan Jumlah Kamar dengan Harga Rumah")
plt.show()
```

```python
# Melatih model dengan Linear Regression
bedrooms = bedrooms.reshape(-1, 1)
linreg = LinearRegression()
linreg.fit(bedrooms, house_price)
```

```python
# Menampilkan plot hubungan antara jumlah kamar dengan harga rumah
plt.scatter(bedrooms, house_price, label="Data Aktual")
plt.plot(bedrooms, linreg.predict(bedrooms), color='red', label="Prediksi")
plt.xlabel("Jumlah Kamar")
plt.ylabel("Harga Rumah ($)")
plt.title("Hubungan Jumlah Kamar dengan Harga Rumah")
plt.legend()
plt.show()
```

```python
# Menampilkan akurasi model
accuracy = linreg.score(bedrooms, house_price)
print(f"Akurasi model (R2 score): {accuracy}")
```

```python
# Prediksi harga rumah dengan data baru
new_bedrooms = np.array([3, 6]).reshape(-1, 1)
predicted_prices = linreg.predict(new_bedrooms)
print(f"Prediksi harga rumah untuk jumlah kamar {new_bedrooms.flatten()}: {predicted_prices}")
```

## **SKLearn Logistic Regression**

Pada latihan ini kita akan menggunakan logistic regression untuk memprediksi apakah seseorang akan membeli setelah melihat iklan sebuah produk.

Tahapan yang dilalui dalam latihan kali ini adalah sebagai berikut:

1. Ubah dataset menjadi Dataframe.
2. Hapus kolom 'User ID'.
3. Pisahkan atribut dan label.
4. Latih model Logistic Regression.
5. Evaluasi akurasi model.

Dataset untuk latihan bisa Anda unduh pada tautan berikut https://www.kaggle.com/datasets/dragonheir/logistic-regression. Seperti biasa, unggah dataset Social_Network_Ads pada session storage Google Colab.

```python
import pandas as pd

# membaca dataset dan mengubahnya menjadi dataframe
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# drop kolom yang tidak diperlukan
data = df.drop(columns=['User ID'])

# jalankan proses one-hot encoding dengan pd.get_dummies()
data = pd.get_dummies(data, dtype="int")
data
```

```python
# pisahkan atribut dan label
predictions = ['Age' , 'EstimatedSalary' , 'Gender_Female' , 'Gender_Male']
X = data[predictions]
y = data['Purchased']
```

```python
# lakukan normalisasi terhadap data yang kita miliki
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns= X.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split

# bagi data menjadi train dan test untuk setiap atribut dan label
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
from sklearn import linear_model

# latih model dengan fungsi fit
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
# uji akurasi model
model.score(X_test, y_test)
```

```python
# Prediksi dengan data baru
new_data = pd.DataFrame({
    'Age': [30, 40],
    'EstimatedSalary': [87000, 120000],
    'Gender_Female': [0, 1],
    'Gender_Male': [1, 0]
})
```

```python
# Menormalkan data baru menggunakan scaler yang sudah fit
scaled_new_data = scaler.transform(new_data)

# Melakukan prediksi
predictions = model.predict(scaled_new_data)
print(f"Prediksi untuk data baru: {predictions}")
```

## **SKLearn K-means**

Tahapan pada latihan kali ini adalah sebagai berikut:

1. Konversi data menjadi Dataframe.
2. Lakukan preprocessing data.
3. Hilangkan kolom 'CustomerID' dan 'gender'.
4. Latih model K-Means.
5. Buat plot untuk Elbow dan Cluster.

Dataset yang akan Anda gunakan adalah data pengunjung sebuah mall fiktif. Dataset bisa Anda dapatkan pada tautan berikut https://www.kaggle.com/datasets/vjchoudhary7/customer-segmentation-tutorial-in-python.

```python
import pandas as pd

#ubah file csv menjadi dataframe
df = pd.read_csv('Mall_Customers.csv')

# tampilkan 3 baris pertama
df.head(3)
```

```python
# ubah nama kolom
df = df.rename(columns={'Gender': 'gender', 'Age': 'age',
                        'Annual Income (k$)': 'annual_income',
                        'Spending Score (1-100)': 'spending_score'})

# ubah data kategorik menjadi data numerik
df['gender'].replace(['Female', 'Male'], [0,1], inplace=True)

# tampilkan data yang sudah di preprocess
df.head(3)
```

```python
from sklearn.cluster import KMeans

# menghilangkan kolom customer id dan gender
X = df.drop(['CustomerID', 'gender'], axis=1)

#membuat list yang berisi inertia
clusters = []
for i in range(1,11):
  km = KMeans(n_clusters=i).fit(X)
  clusters.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# membuat plot inertia
fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
# membuat objek KMeans
km5 = KMeans(n_clusters=5).fit(X)

# menambahkan kolom label pada dataset
X['Labels'] = km5.labels_

# membuat plot KMeans dengan 5 klaster
plt.figure(figsize=(8,4))
sns.scatterplot(x=X['annual_income'], y=X['spending_score'], hue=X['Labels'],
                palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('house_price.csv')

data
```

```python
data.info()
```

```python
data = data.drop(['date', 'yr_renovated', 'street', 'city', 'statezip', 'country', 'yr_built'], axis=1)
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=data)
plt.xticks(rotation=90)
plt.title('Box Plot Before Removing Outliers')
plt.show()
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

outliers = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR)))

data_clean = data[~outliers.any(axis=1)]

print(f'Original dataset shape: {data.shape}')
print(f'Cleaned dataset shape: {data_clean.shape}')
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=data_clean)
plt.xticks(rotation=90)
plt.title('Box Plot After Removing Outliers')
plt.show()
```

```python
X_clean = data_clean.drop(['price'], axis=1)
y_clean = data_clean['price']

X_train_clean, X_test_clean, y_train_clean, y_test_clean = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)
```

```python
rf_model_clean = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model_clean.fit(X_train_clean, y_train_clean)

rf_y_pred_clean = rf_model_clean.predict(X_test_clean)

rf_mae_clean = mean_absolute_error(y_test_clean, rf_y_pred_clean)
rf_mse_clean = mean_squared_error(y_test_clean, rf_y_pred_clean)
rf_rmse_clean = mean_squared_error(y_test_clean, rf_y_pred_clean, squared=False)
rf_r2_clean = r2_score(y_test_clean, rf_y_pred_clean)

print('Random Forest Model Performance after Outlier Removal:')
print(f'Mean Absolute Error: {rf_mae_clean}')
print(f'Mean Squared Error: {rf_mse_clean}')
print(f'Root Mean Squared Error: {rf_rmse_clean}')
print(f'R-squared: {rf_r2_clean}\n')
```

```python
lr_model_clean = LinearRegression()
lr_model_clean.fit(X_train_clean, y_train_clean)

lr_y_pred_clean = lr_model_clean.predict(X_test_clean)

lr_mae_clean = mean_absolute_error(y_test_clean, lr_y_pred_clean)
lr_mse_clean = mean_squared_error(y_test_clean, lr_y_pred_clean)
lr_rmse_clean = mean_squared_error(y_test_clean, lr_y_pred_clean, squared=False)
lr_r2_clean = r2_score(y_test_clean, lr_y_pred_clean)

print('Linear Regression Model Performance after Outlier Removal:')
print(f'Mean Absolute Error: {lr_mae_clean}')
print(f'Mean Squared Error: {lr_mse_clean}')
print(f'Root Mean Squared Error: {lr_rmse_clean}')
print(f'R-squared: {lr_r2_clean}')
```
